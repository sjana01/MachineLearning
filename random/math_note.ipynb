## Pullback Formula

```python
def highestPowerof2(n):
  
    return int(log(n, 2))
    
 
# Driver code
n = 31
print(highestPowerof2(n))
```

```python
def pullback_formula(n,k):

    list_1 = [2**i - 1 for i in range(1,highestPowerof2(n)+1)]
    list_1.insert(0, 1)
    list_2 = [2**i for i in range(0, highestPowerof2(n)+1)]
    
    set_ = [str(i) for i in range(k+1)]
    combinations = [tuples for tuples in itertools.product(set_, repeat = len(list_1))]
    func = lambda x: tuple(int(i) for i in x)
    combinations_int = tuple(func(i) for i in combinations)
    
    set_of_seq = []

    for x in combinations_int:
        if np.sum(np.array(x) * np.array(list_1)) == k  and np.sum(np.array(x) * np.array(list_2)) <= n:
            set_of_seq.append(x)

    return set_of_seq
```

```python
def admissible_sequences(n):
    for i in range(1,n+1):
        print(pullback_formula(n,i))
```

```python
admissible_sequences(7)
```

## Collatz Function

```python
def collatz(n):
    steps = 1
    highest = 1
    while True:
        if n%2 == 0:
            n = n/2
        else:
            n = 3*n + 1
        if n > highest:
            highest = n
        if int(log(n,2)) == log(n,2):
            break
            print(steps)
        steps += 1

    return (steps, highest)
```
